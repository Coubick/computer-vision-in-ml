# №3

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_rectangle(image, top_left, bottom_right, color, thickness=1):
    """
    Рисует прямоугольник на изображении.
    
    Параметры:
    image: входное изображение (numpy array)
    top_left: кортеж (x, y) координат верхнего левого угла
    bottom_right: кортеж (x, y) координат нижнего правого угла
    color: кортеж (B, G, R) или значение для градаций серого
    thickness: толщина линии в пикселях
    """
    result = image.copy()
    
    x1, y1 = top_left
    x2, y2 = bottom_right
    
    height, width = image.shape[:2]
    if thickness != -1:
        for i in range(thickness):
            if x1 + i < width:
                result[y1:y2+1, x1 + i] = color
                
            if x2 - i >= 0:
                result[y1:y2+1, x2 - i] = color
    
        for i in range(thickness):
            if y1 + i < height:
                result[y1 + i, x1:x2+1] = color
                
            if y2 - i >= 0:
                result[y2 - i, x1:x2+1] = color
        
        return result

    else:
        result[y1:y2+1, x1:x2+1] = color
        
        return result

black_image = np.zeros((400, 600, 3), dtype=np.int16)

black_with_rects = draw_rectangle(black_image, (50, 50), (200, 150), (255, 0, 0), thickness=3)
black_with_rects = draw_rectangle(black_with_rects, (250, 50), (400, 150), (0, 255, 0), thickness=-1)
black_with_rects = draw_rectangle(black_with_rects, (450, 50), (550, 150), (0, 0, 255), thickness=15)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(black_image)
plt.title('исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(black_with_rects)
plt.title('изменённое изображение')
plt.axis('off')

plt.tight_layout()
plt.show()
```
